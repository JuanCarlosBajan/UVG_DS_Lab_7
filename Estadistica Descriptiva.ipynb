# 1. Estadísticas descriptivas básicas:

```python
import pandas as pd

# Crear un DataFrame
data = pd.read_csv('diabetes.csv') 

# Estadísticas descriptivas
print(data.describe())
```

# 2. Visualización de la distribución de las variables:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogramas para cada variable
data.hist(figsize=(10,10))
plt.tight_layout()
plt.show()

# Boxplots para identificar posibles valores atípicos
plt.figure(figsize=(12,8))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Verificar la presencia de valores nulos o atípicos:

```python
# Mostrar la cantidad de valores nulos por columna
print(data.isnull().sum())
```
